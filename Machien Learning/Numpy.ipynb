***Broadcasting***

Broadcasting is a powerful mechanism that allows NumPy to perform operations on arrays of different shapes without needing to create unnecessary copies of data. It works by "stretching" the smaller array to match the shape of the larger one, so that element-wise operations can be performed efficiently.

Instead of explicitly reshaping arrays to match each other, broadcasting automatically handles the differences in dimensions. However, it follows a specific set of rules to determine if broadcasting can be applied.


```python
import numpy as np

matrix = np.array([[1, 2, 3], [4, 5, 6]])  # Shape (2, 3)
scalar = 2  # Shape ()

# Broadcasting allows us to multiply each element of the matrix by the scalar
result = matrix * scalar
print(result)
```
