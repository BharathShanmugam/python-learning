```python
b=a**4
```

```python
x=np.arange(0,11)
```

```python
x
```

```python
y=x*2
```

```python
fig=plt.figure()
axies=fig.add_axes([0,0,1,1])
axies.plot(a,b)
axies.plot(x,y)
axies.legend_()
```

```python
import matplotlib.pyplot as plt



x = [1, 2, 3, 4, 5]

y = [10, 20, 25, 30, 40]



plt.plot(x, y)

plt.xlabel('X-axis')

plt.ylabel('Y-axis')

plt.title('Simple Line Plot')
plt.savefig("myfirstmat.jpg")
plt.show()
```

```python

import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [2, 5, 8, 10]

fig=plt.figure(figsize=(10,5),dpi=200)
plt.bar(x, y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('Data Visualization in Python')
plt.savefig("Linked.jpg")
```
