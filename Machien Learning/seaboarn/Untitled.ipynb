***1. Scatter Plot***
    • Scenario: Imagine you're a kid interested in knowing if people with bigger feet are taller. You have a list of your friends' foot sizes and their heights.
    • Use: Scatter plots help you see if there's a relationship between two things, like foot size and height.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data: friends' foot sizes and heights
friends = {'foot_size': [5, 6, 7, 8, 9], 'height': [100, 110, 120, 130, 140]}
df = pd.DataFrame(friends)

# Scatter plot: Foot size vs Height
sns.scatterplot(x='foot_size', y='height', data=df)
plt.title('Foot Size vs Height')
plt.show()
```

Explanation: Each dot shows a friend’s foot size and height. You can see if bigger foot sizes are related to taller heights.
