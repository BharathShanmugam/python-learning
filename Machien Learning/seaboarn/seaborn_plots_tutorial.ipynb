## 1. Scatter Plot

**Scenario:** Imagine you're a kid interested in knowing if people with bigger feet are taller. You have a list of your friends' foot sizes and their heights.

**Use:** Scatter plots help you see if there's a relationship between two things, like foot size and height.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data: friends' foot sizes and heights
friends = {'foot_size': [5, 6, 7, 8, 9], 'height': [100, 110, 120, 130, 140]}
df = pd.DataFrame(friends)

# Scatter plot: Foot size vs Height
sns.scatterplot(x='foot_size', y='height', data=df)
plt.title('Foot Size vs Height')
plt.show()
```

**Explanation:** Each dot shows a friendâ€™s foot size and height. You can see if bigger foot sizes are related to taller heights.



## 2. Distribution Plot

**Scenario:** You want to know how often different grades (A, B, C, etc.) are given out in class.

**Use:** Distribution plots show how things (like grades) are spread out in a group.

```python
# Sample data: Grades of students
grades = {'grades': ['A', 'A', 'B', 'C', 'B', 'A', 'C', 'B', 'A', 'B']}
df = pd.DataFrame(grades)

# Distribution plot: Grades
sns.histplot(df['grades'], kde=False)
plt.title('Grade Distribution')
plt.show()
```

**Explanation:** This shows how many of your classmates got each grade. If there are lots of A's, it means most students did well.



## 3. Categorical Plot

**Scenario:** You and your friends have a lemonade stand. You want to know how much money you make on weekdays vs weekends.

**Use:** Categorical plots compare different groups, like weekdays and weekends.

```python
# Sample data: Money earned on different days
sales = {'day': ['Monday', 'Tuesday', 'Saturday', 'Sunday'], 'earnings': [20, 25, 50, 45]}
df = pd.DataFrame(sales)

# Bar plot: Earnings by day
sns.catplot(x='day', y='earnings', kind='bar', data=df)
plt.title('Lemonade Stand Earnings by Day')
plt.show()
```

**Explanation:** This bar chart shows how much you earned each day. It helps you compare and see if you make more money on weekends.



## 4. Box Plot

**Scenario:** Letâ€™s say youâ€™re curious about how much pocket money your classmates get. You want to know if anyone gets much more or much less than others.

**Use:** Box plots show the range of data and if there are any 'outliers' (people who get much more or much less).

```python
# Sample data: Classmates' pocket money
pocket_money = {'name': ['John', 'Sara', 'Mike', 'Anna', 'Tom'], 'money': [5, 10, 15, 20, 50]}
df = pd.DataFrame(pocket_money)

# Box plot: Pocket money distribution
sns.boxplot(y='money', data=df)
plt.title('Pocket Money Distribution')
plt.show()
```

**Explanation:** The box shows most kids' pocket money, while the dots (if any) show the kids who get much more or much less.



## 5. Comparison Plots

**Scenario:** You and your friends track how many pages you read each week to see who reads the most over a month.

**Use:** Comparison plots help you compare trends or changes over time.

```python
# Sample data: Pages read by you and your friend over 4 weeks
reading = {'Week': [1, 2, 3, 4], 'You': [10, 15, 20, 25], 'Friend': [12, 17, 22, 28]}
df = pd.DataFrame(reading)

# Line plot: Compare pages read by week
sns.lineplot(x='Week', y='You', data=df, label='You')
sns.lineplot(x='Week', y='Friend', data=df, label='Friend')
plt.title('Pages Read Over 4 Weeks')
plt.legend()
plt.show()
```

**Explanation:** You can see who read more each week and whether both of you are reading more as time goes on.



## 6. Grid Plots

**Scenario:** Youâ€™re curious about how different things are related, like how many hours of TV you watch and your grades.

**Use:** Grid plots let you look at multiple relationships at once.

```python
# Sample data: Study hours, TV hours, and grades
students = {'study_hours': [1, 2, 3, 4, 5], 'tv_hours': [5, 4, 3, 2, 1], 'grades': [60, 70, 80, 90, 100]}
df = pd.DataFrame(students)

# PairGrid plot: Study hours, TV hours, and grades
g = sns.PairGrid(df)
g.map(sns.scatterplot)
plt.show()
```

**Explanation:** This grid shows how studying, watching TV, and grades relate. You can spot patterns across different combinations of things.



## 7. Matrix Plot (Heatmap)

**Scenario:** You and your friends play a game where you guess each other's favorite colors. You want to know how often each color is guessed correctly.

**Use:** Matrix plots (or heatmaps) show how much things are related, often with colors to make it easy to spot trends.

```python
# Sample data: Favorite color guessing game
data = {'Red': [5, 3, 1], 'Blue': [2, 4, 3], 'Green': [0, 1, 5]}
df = pd.DataFrame(data, index=['John', 'Anna', 'Tom'])

# Heatmap: Guesses of favorite colors
sns.heatmap(df, annot=True, cmap='coolwarm')
plt.title('Guessing Favorite Colors')
plt.show()
```

**Explanation:** The heatmap shows how well each friend guessed othersâ€™ favorite colors. Darker or lighter colors show more or fewer correct guesses.
