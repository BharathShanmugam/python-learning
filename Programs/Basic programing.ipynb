#### Breakdown of the Loops

- **Outer Loop: `for (int i = 0; i < size; i++)`**
  - Each pass through the array.

- **Inner Loop: `for (int j = 0; j < size - i - 1; j++)`**
  - Compares adjacent elements and swaps if necessary.

#### Output
- Prints the sorted array.
