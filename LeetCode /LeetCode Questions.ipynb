# Merge Two Strings Alternately

````python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        i, j = 0, 0  # Initialize indices for both strings
        res = []     # Initialize an empty list to hold the result

        while i < len(word1) and j < len(word2):
            res.append(word1[i])  # Append character from word1
            res.append(word2[j])  # Append character from word2
            i += 1  # Move to the next character in word1
            j += 1  # Move to the next character in word2
        
        # Append remaining characters from either string
        res.append(word1[i:])  # Remaining characters in word1
        res.append(word2[j:])  # Remaining characters in word2
        
        return "".join(res)     # Combine the list into a string
````

````python
class Solution:
    def gcdOfStrings(self, str1: str, str2: str) -> str:
        # Determine the lengths of the two strings
        len1 = 0
        for _ in str1:
            len1 += 1
            
        len2 = 0
        for _ in str2:
            len2 += 1

        # Find the smaller length manually
        min_length = len1 if len1 < len2 else len2
        
        # Check all possible lengths for the common substring
        for length in range(min_length, 0, -1):
            # Check if the current length can divide both strings
            if len1 % length == 0 and len2 % length == 0:
                # Get the potential common substring
                substring = ""
                for i in range(length):
                    substring += str1[i]
                
                # Check if repeating the substring forms both str1 and str2
                repeated_str1 = ""
                for i in range(len1 // length):
                    repeated_str1 += substring
                
                repeated_str2 = ""
                for i in range(len2 // length):
                    repeated_str2 += substring
                
                if repeated_str1 == str1 and repeated_str2 == str2:
                    return substring
        
        # If no common divisor is found, return an empty string
        return ""

# Example usage:
solution = Solution()
print(solution.gcdOfStrings("ABCABC", "ABC"))    # Output: "ABC"
print(solution.gcdOfStrings("ABABAB", "ABAB"))   # Output: "AB"
print(solution.gcdOfStrings("LEET", "CODE"))     # Output: ""
````
